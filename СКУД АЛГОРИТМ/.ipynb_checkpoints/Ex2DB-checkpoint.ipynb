```python
import pandas as pd
from datetime import datetime, timedelta
from sqlalchemy import create_engine
```

```python
import os
import pandas as pd

folder_path = r'C:\Users\shaih\OneDrive\Рабочий стол\КТЖ\КТЖ СКУД\СКУД данные'
dataframes = []

# Рекурсивный поиск Excel-файлов
for root, dirs, files in os.walk(folder_path):
    for file in files:
        if file.endswith(('.xlsx', '.xls')):
            file_path = os.path.join(root, file)
            df = pd.read_excel(file_path)
            dataframes.append(df)
            print(f"Файл: {file_path}")
            print(df.head())  # Печатаем первые строки для проверки

# Объединение всех DataFrame в один
combined_df = pd.concat(dataframes, ignore_index=True)

# Сохранение в новый файл (опционально)
combined_df.to_excel(r'C:\Users\shaih\OneDrive\Рабочий стол\КТЖ\КТЖ СКУД\combined_data.xlsx', index=False)

print("\nОбъединённый DataFrame:")
print(combined_df.head())
```

```python
DATABASE_URL = "postgresql+psycopg2://postgres:2340176@localhost:5432/scud_db"
engine = create_engine(DATABASE_URL)
```

```python
combined_df = combined_df.rename(columns={
    "ID сотрудника": "employee_id",
    "Имя": "first_name",
    "Фамилия": "last_name",
    "Имя отдела": "department",
    "Название точки учёта": "checkpoint_name",
    "время посещения": "visit_time"
})
```

```python
# Запись в PostgreSQL
# if_exists="append" → дописывать в таблицу, не удаляя старые данные
combined_df.to_sql("scud_data", engine, if_exists="append", index=False)

print("\n✅ Все данные успешно загружены в PostgreSQL!")
```
